```python
import numpy as np
```

```python
path = '../data/raw/AB_NYC_2019.csv'
rows = []
with open(path, 'r') as file:
    data = file.readlines()
    for lines in data:
        lines = lines.strip()
        inside = False
        cols = []
        col_current = ""
        for line in lines:
            if line == '"':
                inside = not inside
            elif line == ',' and not inside:
                cols.append(col_current)
                col_current = ""
            else:
                col_current += line
            cols.append(col_current)
        rows.append(cols)
data = np.array(rows, dtype=object)
```
